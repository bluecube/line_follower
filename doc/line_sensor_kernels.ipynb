Exploring cross corelation kernels for detecting line in the 8x1 image from the line sensor

```python
import numpy
import matplotlib.pyplot
```

```python
def try_kernel(kernel):
    background_noise = numpy.random.randint(0, 32, 8, numpy.uint8)
    
    try_pair("strong_peak", background_noise + [0, 0, 0, 200, 200, 0, 0, 0], kernel)
    try_pair("blurred_peak", background_noise + [0, 0, 0, 150, 200, 150, 0, 0], kernel)
    try_pair("two_peaks", background_noise + [0, 150, 200, 150, 0, 200, 200, 0], kernel)
    try_pair("almost_missed", background_noise + [0, 0, 0, 0, 0, 0, 0, 150], kernel)
    try_pair("too_wide", background_noise + [0, 0, 100, 150, 150, 100, 0, 0], kernel)
    try_pair("inverse", 255 - (background_noise + [0, 0, 0, 200, 200, 0, 0, 0]), kernel)
```

```python
def try_pair(name, a, kernel):
    fig, (original_axes, correlated_axes) = matplotlib.pyplot.subplots(1, 2)
    fig.suptitle(name)
    original_axes.bar(numpy.arange(len(a)), a)
    correlated = numpy.correlate(a, kernel, mode="same")
    correlated_axes.bar(numpy.arange(len(correlated)), correlated)
```

```python
try_kernel([-1, 1, 1, -1])
```
